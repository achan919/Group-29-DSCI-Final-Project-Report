```r
# load in libraries
library(tidyverse)
library(tidymodels)
```

```r
# load in data files
players <- read_csv("https://raw.githubusercontent.com/achan919/dsci-final-project/refs/heads/main/players.csv")
sessions <- read_csv("https://raw.githubusercontent.com/achan919/dsci-final-project/refs/heads/main/sessions.csv")
```

```r
# tidy the data

# change the data type for 
players <- mutate(players, experience = as.factor(experience))
players <- mutate(players, gender = as.factor(gender))

# change the time values for the original start time and original end time to be more readable
sessions <- sessions |>
    mutate(original_end_time = as.numeric(as.POSIXct(end_time, format = "%d/%m/%Y %H:%M")),
           original_start_time = as.numeric(as.POSIXct(start_time, format = "%d/%m/%Y %H:%M")))
```
